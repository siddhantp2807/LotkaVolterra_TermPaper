```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
from scipy.integrate import odeint
```

```python
x, y = np.meshgrid(np.arange(0, 3, 0.05), np.arange(0, 3, 0.05))

alpha, beta = 1, 2
delta, gamma = 3, 4

xdot = alpha*x - beta*x*y
ydot = delta*x*y - gamma*y

pl.streamplot(x, y, xdot, ydot, density = 1.2, linewidth = 0.8)
# pl.plot(delta/gamma, alpha/beta, 'o')
pl.xlabel(r"$\dot{x}$")
pl.ylabel(r"$\dot{y}$")
pl.title("Phase Portrait of one prey-one predator model")
pl.savefig("OnePredatorOnePrey.png", dpi=240)
```

```python
def model(z, t, a, b, c, d) :
    x,y = z[0], z[1]
    dxdt = a*x - b*x*y # Prey
    dydt = c*x*y - d*y # Predator

    return [dxdt, dydt]
```

```python
t = np.linspace(0, 10, 1000)
x = np.empty_like(t)
y = np.empty_like(t)
```

```python
z0 = [2, 5]
for i in range(1, len(t)):
    tspan = [t[i-1], t[i]]
    z = odeint(model, z0, tspan, args=(1, 2, 4, 5))
    x[i], y[i] = z[1][0], z[1][1]
    z0 = z[1]
```

```python
plt.plot(t, x, 'k-', label="Prey Population")
plt.plot(t, y, 'b-', label="Predator Population")
plt.legend(loc='best')
plt.title("Population plot of Predator and Prey")
plt.xlabel("time")
plt.savefig("OnePredatorOnePreyXY.png", dpi=240)
```
